```python
import numpy as np
import pandas as pd
```

```python
car = pd.read_csv('car.data')
car
```

```python
car.describe()
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(car, test_size=0.1, random_state=42)
```

```python
train_set
```

```python
test_set
```

```python
train_features = train_set.drop('target', axis = 1)
train_lables = train_set['target'].copy()

test_features = test_set.drop('target', axis =1)
test_lables = test_set['target'].copy()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth=9, random_state=0)
model.fit(np.array(train_features), np.array(train_lables))
test_features_arr = np.array(test_features)
predictions = model.predict(test_features_arr)
predictions
```

```python
test_lables_arr = np.array(test_lables)
test_lables_arr
```

```python
count = 0
for _ in range(len(predictions)):
    if(predictions[_] != test_lables_arr[_]):
        count = count + 1
count
```
